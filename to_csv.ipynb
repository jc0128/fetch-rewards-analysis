```python
### Importing Packages 
import pandas as pd
import numpy as np
import json
from pandas import json_normalize
```

```python
### Reading in JSON Files

#### Put your File path here to read in json file:
json_file_path = '/Users/name/Documents/FolderName/receipts.json'

# Read JSON file 
data = []
with open(json_file_path, 'r') as f:
    for line in f:
        data.append(json.loads(line.strip())) 
```

```python
### ONLY USED To handle receipts portion of the json
df_receipts = pd.DataFrame(data)
df_receipts = pd.json_normalize(data, sep='_', errors="ignore")

df_receipts.rename(columns={"_id_$oid": "receipt_id"}, inplace=True)
df_receipts.head()

```
